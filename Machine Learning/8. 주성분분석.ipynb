차원이 증가하면 데이터를 잘 설명하지 못하고 독립변수끼리 상관관계가 높다면 다중공선성이 높아 모델의 성능이 저하된다.

-> 차원을 줄이면서, 동시에 정보의 손실을 최소화하는 방법이 PCA(주성분분석)이다.

행렬을 이용하면 여러 개의 확률변수가 서로 어떤 관계를 가지는지를 쉽게 표현할 수 있다.

특히, 각 데이터의 분산과 공분산을 이용해 만드는 공분산 행렬이 이에 해당한다.

공분산 : 2개의 확률변수의 선형 관계를 나타내는 값

공분산 연산 전에는 표준화가 필요하다.

주성분 분석에서 고유값(Eigenvalue)은 공분산 행렬 또는 상관계수 행렬의 고유값으로서, 주성분들이 원본 데이터의 분산을 설명하는 정도를 나타냅니다.

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler

#한글과 음수 표시
matplotlib.rc('font',family='Malgun Gothic')
matplotlib.rc("axes",unicode_minus = False)

#주성분 분석
from sklearn.decomposition import PCA

#회귀 분석
import statsmodels.api as sm

import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./실습화일/2. Big Data 분석/직무능력평가.csv",encoding="euc-kr")

#데이터 분리(Y,X) -> 설명변수를 이용한 주성분 분석
df_x = df.drop("EVAL",axis=1)
df_y = df["EVAL"]

display(df_x.head())
display(df_y.head())
```

변수들의 선형관계를 기반으로 주성분 분석을 진행하기 때문에 결과의 신뢰도를 높인다.

선형성이 낮은 데이터의 경우에는 다른 비선형 차원 축소 기법이나 비선형 모델을 고려하는 것이 좋을 수 있습니다.

```python
# 산점도 행렬
sns.pairplot(df_x)

# Pearson 상관계수: pearson,kendall,spearman
df_x.corr().round(3)
```

DISSATIS는 CHANCE, ADVANCE와 상관관계가 높음.
CHANCE는  ADVANCE와 상관관계가 높음.

```python
# 요약통계량 -> scale변환 필요 확인
display(df_x.describe().round(3))

#변수 이름 저장
x_cols = df_x.columns

# scale 변환: ndarray 형식
scaler = StandardScaler()
nd_scaled = scaler.fit_transform(df_x)

#변환된 데이터 및 형태 확인
print("Scaled Data :\n {}".format(nd_scaled[:5].round(3)))
print("Shape :\n {}".format(nd_scaled.shape))
```

```python
#DataFrame으로 저장
df_scaled = pd.DataFrame(nd_scaled, columns=x_cols)
display(df_scaled.head().round(3))

#Scaled 변환 데이터 요약 통계량
df_scaled.describe().round(3)
```

```python
# 주성분 분석:주성분1(Prin1) ~ 주성분6(Prin6) 생성 및 저장
pca = PCA(n_components=6)
nd_pca = pca.fit_transform(df_scaled)

#생성된 주성분 점수 확인
print("PCA Data :\n {}".format(nd_pca[:5]))

#생성된 주성분 점수 Row, Column 개수 확인
print("Shape : \n {}".format(nd_pca.shape))
```

```python
# 주성분 점수 저장(Prin1 ~ Prin6, DataFrame형식) 및 확인
df_pca = pd.DataFrame(nd_pca,columns=["Prin1","Prin2","Prin3","Prin4","Prin5","Prin6"])

# 주성분 분석용 데이터와 주성분 점수 결합
df_scaled_pca = df_scaled.join(df_pca)
df_scaled_pca.head()
```

```python
# 주성분 분석 통계량 저장: 주성분 번호, Eigenvalue,Eigenvalue ratio
df_pca_stat = pd.DataFrame()
df_pca_stat["PrinNo"] = [l for l in range(1,7)]
df_pca_stat["EigenValues"] = pca.explained_variance_   # 주성분 고유값
df_pca_stat["EigenValueRatio"] = pca.explained_variance_ratio_   # 고유값 비율
df_pca_stat["CumEigenValueRatio"] = np.cumsum(pca.explained_variance_ratio_)  # 고유값 비율의 누적합
df_pca_stat.round(3)
```

적정 주성분 선택

: 상관계수 기준(고유값>1) 2개, 공분산 기준(고유값 누적비율>0.8) 3개

```python
#Scree Plot
plt.plot(df_pca_stat["PrinNo"],df_pca_stat["EigenValues"],label="고유값")
#그래프 label,legend 설정
plt.ylabel("고유값(EigenValue)")
plt.xlabel("주성분 번호")
plt.legend(loc="best")
```

```python
# 고유값 비율 그래프 생성
plt.plot(df_pca_stat["PrinNo"],df_pca_stat["EigenValueRatio"],label="고유값")
# 누적된 고유값 비율 그래프 생성
plt.plot(df_pca_stat["PrinNo"],df_pca_stat["CumEigenValueRatio"],label="누적 고유값")

plt.ylabel("고유값 비율")
plt.xlabel("주성분 번호")
plt.legend(loc="best")
```

```python
# 주성분 분석 데이터의 변수 이름
x_cols = pd.DataFrame({"Feature": df_x.columns})

#주성분 Prin1~6까지의 고유벡터 저장
# 전치 결과 pd.DataFrame(pca.components_,columns=["Prin1","Prin2","Prin3","Prin4","Prin5","Prin6"],axis=1)
# 파이썬 PCA 분석 시 SVD 연산 방식의 차이 반영(전치,호 변경)
df_eigenvector = pd.concat([x_cols,pd.DataFrame(pca.components_.T * -1, columns=["Prin1","Prin2","Prin3","Prin4","Prin5","Prin6"])],axis=1)
df_eigenvector.round(3)
```

고유값 확인 및 주성분해석

: 1주성분은 배움 기회와 승진 간의 양의 연관성
, 2주성분은 불만처리와 특권 허용 간의 음의 연관성 확인

```python
# 산점도 행렬
sns.pairplot(df_pca)
df_pca.corr().round(3)
```

주성분은 서로 상관계수가 0 이므로 독립적이라고 할 수 있다.
즉, 서로 독립변수 간의 중복성이 없다는 것을 의미

```python
#다중선형 회귀 분석: "원천 데이터 활용"
# 원천 데이터에 절편(const) 추가       # 절편은 항상 추가되어야 한다
df_x_const = sm.add_constant(df_x)

#회귀모델 생성
reg_model = sm.OLS(df_y,df_x_const)
#회귀모델 적합
reg_model = reg_model.fit()

#모델 성능 확인
print(reg_model.summary())
```

```python
# 참조 
# 다중선형 회귀 분석: Scale 변환 데이터 활용
# Scale 변환 데이터에 절편(const) 컬럼 추가
df_scaled_const = sm.add_constant(df_scaled)

#회귀모델 생성
reg_model = sm.OLS(df_y,df_scaled_const)
#회귀모델 적합
reg_model = reg_model.fit()

#모델 성능 확인
print(reg_model.summary())
```

```python
#다중선형 주성분 회귀분석: 주성분 점수(score) 데이터 활용
# Eigenvalue, Scree plot 기준으로 적정 주성분 3개 선택
x_var= ["Prin1","Prin2","Prin3"]

#선택된 주성분 점수 데이터에 절편(const)컬럼추가
df_pca_const = sm.add_constant(df_pca[x_var])

#회귀모델 생성
reg_model=sm.OLS(df_y,df_pca_const)

#회귀모델 적합
reg_model = reg_model.fit()
#모델 성능 확인
print(reg_model.summary())
```

전체 설명변수를 이용한 모델보다 3개 주성분을 이용한 회귀모델의 설명력이 약 11% 저하됨 (주성분 4~6을 제외하였기 때문에)

주성분 회귀분석의 회귀계수는 모두 유의
