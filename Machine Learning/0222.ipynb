```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K0','K0','K0'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
result = pd.merge(left,right, on = ['key1','key2'])
print(result)

result = pd.merge(left, right, on = ['key1','key2'],how='left')
print(result)

result = pd.merge(left, right, on = ['key1','key2'],how='right')
print(result)

result = pd.merge(left, right, on = ['key1','key2'],how='outer')
print(result)

result = pd.merge(left, right, on = ['key1','key2'],how='outer', indicator = True)
print(result)
```

```python
left = pd.DataFrame({'key1':[1,2,3,4],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'C':['C1','C2','C5'],
                   'D':['D1','D2','D5']},
                    index=[1,2,5])

result = left.join(right)  # 인덱스 기반
print(result)

result = left.join(right, how = 'right')
print(result)

result = left.join(right, how='inner')
print(result)

result = left.join(right, how='outer')
print(result)

result = left.join(right, on='key1')
print(result)

result2 = pd.merge(left, right, left_on='key1', right_index=True, how = 'left') # 지정 열 기반
print(result2)
```

```python
df = pd.DataFrame({'index':[1,1,1,2,2,2],
                  'column':['col_1','col_2','col_3','col_1','col_2','col_3'],
                  'value_1':[1,2,3,4,5,6],
                  'value_2': ['x','y','z','q','w','t']})

print(df)

print(df.pivot(index = 'index', columns = 'column', values=['value_1','value_2']))
print(df.pivot(index = 'index', columns = 'column')['value_1'])


print(df.pivot(index = 'index', columns = 'column', values=['value_1','value_2'])) #복수의 pivot column 지정
print(df.pivot(index = 'index', columns = 'column')) #미지정
print(df.pivot(columns='column', values='value_1')) #인덱스 미 지정 시 변환되는 데이터 구조 주의

print(df.melt(id_vars=['index','column'])) #melt: id_vars를 제외한 나머지 컬럼을 풀어서 쌓음
print(df.melt(id_vars=['index','column'],value_vars=['value_1']))
```

```python
df = pd.DataFrame({'index':[1,2,3],'alpha':['A','B','C'],'values':[11,22,33],'xyz':['x','y','z']})
print(df)

df_stack = df.stack() #컬럼의 인덱스를 행의 여러 개 label 중 가장 왼쪽의 label로 un-pivot하여 변환
print(df_stack)
print("stack 데이터 크기: ",df_stack.shape)

df_unstack = df_stack.unstack() #stack의 반대
print(df_unstack)
```

```python
df = pd.DataFrame({'factor1':['A','A','A','A','A','B','B','B','B'],
                  'factor2':['a','b','c','a','b','a','b','c','c'],
                  'value':[11,13,12,12,14,12,13,11,12]})

print(df)

print(pd.crosstab(df.factor1, df.factor2))
print(pd.crosstab(df.factor1,df.factor2,rownames=['Fact1'], colnames=['Fact2'], margins=True))
print(pd.crosstab(df.factor1,[df.factor2, df.value]))
print(pd.crosstab(df.factor1, df.factor2,margins=True, normalize=True))
```

```python
pop_seoul = pd.read_csv("/home/piai/python/2주차/실습화일/1 Big Data 분석 기초/서울인구통계.txt",sep="\t",
                        thousands=',',skiprows=2, encoding = "utf-8")
print(pop_seoul.head())

pop1 = pop_seoul[['자치구','계','계.1','계.2','65세이상고령자']]
pop_seoul = pop1.drop([0])
print(pop_seoul.head())

pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                         pop_seoul.columns[1]:'인구수',
                         pop_seoul.columns[2]:'한국인',
                          pop_seoul.columns[3]:'외국인',
                         pop_seoul.columns[4]: '고령자'}, inplace=True)

print(pop_seoul.head())
print()

pop_seoul['외국인 비율']=pop_seoul['외국인']/pop_seoul['인구수']*100
pop_seoul['고령자 비율']=pop_seoul['고령자']/pop_seoul['인구수']*100

pop_seoul.head()

print(pop_seoul.sort_values(by='외국인 비율',ascending=False).head())
print(pop_seoul.sort_values(by='고령자 비율',ascending=False).head())
```
