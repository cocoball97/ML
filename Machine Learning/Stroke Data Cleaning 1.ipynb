# 5.1.13 t-검정

```python
df.info()
```

```python
from scipy import stats

data_1 = df[df['Age'] == '노게']['Length']
data_0 = df[df['Age'] == '성게']['Length']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['Age'] == '노게']['Diameter']
data_0 = df[df['Age'] == '성게']['Diameter']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['Age'] == '노게']['Height']
data_0 = df[df['Age'] == '성게']['Height']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['Age'] == '노게']['Weight']
data_0 = df[df['Age'] == '성게']['Weight']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['Age'] == '노게']['Shucked Weight']
data_0 = df[df['Age'] == '성게']['Shucked Weight']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['Age'] == '노게']['Viscera Weight']
data_0 = df[df['Age'] == '성게']['Viscera Weight']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['Age'] == '노게']['Shell Weight']
data_0 = df[df['Age'] == '성게']['Shell Weight']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```
