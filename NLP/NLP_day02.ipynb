```python
from sklearn.feature_extraction.text import CountVectorizer
#CountVectorizer:단어의 빈도수로 벡터 생성
```

```python
corpus=['나는 지금 자연어 처리를 공부해요. 나는 자연어 좋아요.']
vector=CountVectorizer()
```

```python
vector.fit_transform(corpus)
#fit:각각의 단어별로 인덱스 부여(0~5)
#transform: corpus에 저장된 문장을 fit 결과에 맞춰 변환(빈도수)해라=> [2 1 2 1 1 1]
# sparse matrix(희소행렬:행렬 요소값이 대부분 0인 행렬=>공간낭비,연산량이 많음)
# [
#     [2 1 2 1 1 1]
# ]
vector.fit_transform(corpus).toarray()
```

```python
# 코사인 유사도를 사용해 DTM (유사한 문서 찾기, But, 단순 많이,적게 등장했다고 해서 중요한 단어일까??)
# > tfidf 변환(특정 단어가 특정 문서에서 자주 등장하고 전체 문서 집합에서 희귀할수록 높은 값 => 해당 문서에서 중요 )

#  DTM : 텍스트 데이터를 수치적인 형태로 변환 - CountVectorizer 사용하여 단어의 빈도수 확인
# 문서 유사도 : 얼마나 많은 단어가 공통적으로 들어있는지

# tf(d,t) : 문서 d에서 단어 t의 등장 횟수(DTM에서 행렬의 요소값이 tf임)
# df(t) : 단어 t가 등장한 문서의 개수
# idf(d,t) : log(n / (df(t)+1))
# - n은 전체 문서의 개수
# - log

# ex) n=1000

# idf(d,t)=log(1000 / (df(t)+1))
# df(t) =1, idf(d,t)는 log(1000/1)은 3
# df(t) =10, idf(d,t)는 log(1000/10)은 2

# tfidf=tf*idf
# 특정 단어가 많이 등장하면서, 흔치 않은 단어여야 함
# 특정 단어가 특정 문서에서 자주 등장하고 전체 문서 집합에서 희귀할수록 높은 값을 갖게 됩니다.
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm
```

```python
dot(d3, d2) / (norm(d3)*norm(d2)) #1
```

```python
# 사실 이 함수만 이해하면 간단한 추천 시스템 끝!!! 찬찬히 살펴보아요
def recommend(t, cosine_sim=cos_sim): #t에는 제목 전달
    idx=data[t]
    sim_scores=list(enumerate(cosine_sim[idx]))
    # print(sim_scores)
    ss=sorted(sim_scores, key=lambda x:x[1], reverse=True) # 연관성이 높다고 생각되는 상위 10개만 추출
    ss=ss[1:11]
    print(ss)
    res=[i[0] for i in ss]
    print(res)
    return df['title'].iloc[res]   # iloc 을 이용해 모든 행에 조건 인덱스만 가져오기
```

```python
recommend('Toy Story')
```
